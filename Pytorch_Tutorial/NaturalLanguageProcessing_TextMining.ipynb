## NLTK function
- **Tokenization** : minimal piece of text(machine understand) == Token, Breakdown to sentences == (nltk.sent_tokenize)
- **word_tokenize** == split(), regexp.tokenize(pattern="\w+") == "!." in token
- **Stemming** : Porter, Snowball, Lancaster (decide on me)
  - stem ex)'cutting' to 'cut'
- **TextNormalization**
- **Lemmatization** : considers the context converts the word to its meaningful base form
  - lemmatization ex) 'Caring' to 'Care'/ 'feet' to 'foot'
- **StopWordRemoval** : do not add any meaningful value to the extraction (background noise)
  - stopword ex) 'a','of','and','that are not' etc..
- **Replacing and Correcting words**
  - ex) upper to lower(lower())
- **Removing numbers** (import re)
- **Removing punctuation**...(import string)
- **Removing whitespaces** (import string)
- **Sentimental Analysis** : to guage th sentiments, attitudes, emotions of writer of speaker depending on the computational treatment
  - VADER

# Tokenize

```python
from nltk.tokenize import sent_tokenize
myString=df_copy['review']
print(myString)
```

```python
# 0행 토큰화
myString0=myString[0]
print(sent_tokenize(myString0))
```

```python
# 단어 토큰화, 특수문자 제거, 숫자 추출
from nltk.tokenize import word_tokenize, regexp_tokenize
print(word_tokenize(myString0))
print(regexp_tokenize(myString0,pattern="\w+"))
print(regexp_tokenize(myString0,pattern="\d+"))
```

# Stemming & Text Normalization
  - cutting down a token to its root stem

## PorterStemmer

```python
from nltk.stem import PorterStemmer
porter=PorterStemmer()
print(porter.stem('trying')) #try
print(porter.stem('rescued')) #rescue
print(porter.stem('forced')) #force
```

## LancasterStemmer

```python
from nltk.stem import LancasterStemmer
lan=LancasterStemmer()
print(lan.stem('trying')) #try
print(lan.stem('rescued')) #rescue
print(lan.stem('forced')) #force
```

## SnowballStemmer

```python
from nltk.stem import SnowballStemmer
snow=SnowballStemmer("english")
print(snow.stem('trying')) #try
print(snow.stem('rescued')) #rescue
print(snow.stem('forced')) #force
```

# Lemmatization

```python
import nltk
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
lemm=WordNetLemmatizer()
print(lemm.lemmatize('bats')) #bat
print(lemm.lemmatize("tries")) #try
print(lemm.lemmatize('rescued')) #rescue
print(lemm.lemmatize('forced')) #force
```

# Stop Word Removal

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
stop=stopwords.words('english')
postPa=[x for x in myString0.split() if x not in stop]
print(postPa)
```

# Replacing and correcting words
- converting text to either lower to upper
- removing numbers, punctuation, whitespace
- finding synonyms & anotnyms

```python
import re #숫자 삭제
print(re.sub(r'\d+',"",myString0))
```

```python
import string # 특수문자 삭제
print(myString0.translate(str.maketrans("","",string.punctuation)))
```

```python
import string # 앞 뒤 공백 삭제
print(myString0.strip())
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyser=SentimentIntensityAnalyzer()
```

```python
count=0
for i in myString:
  
  polarity=analyser.polarity_scores(i)
  pos=polarity["pos"]
  neu=polarity["neu"]
  neg=polarity["neg"]
  count+=1

  print(f'positive sentence in "{count}" is : {round(pos*100,2)}%' )
  print(f'neutral sentence in "{count}" is : {round(neu*100,2)}%' )
  print(f'negative sentence in "{count}" is : {round(neg*100,2)}%' )
```
